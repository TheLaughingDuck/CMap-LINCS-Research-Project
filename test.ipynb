## Compoundinfo_beta.txt
_Metadata for cell lines (note there is one entry per compound/moa/target/structure combination, so some compounds appear more than once)._

* **pert_id**: A unique identifier for a perturbagen that refers to the perturbagen in general, not to any particular batch or sample. Can be considered the name of the perturbagen/compound. Likely not useful as a predictor.

* **cmap_name**: The internal (CMap-designated) name of a perturbagen. By convention, for genetic perturbations CMap uses the HUGO gene symbol. A slightly less formal name. For example "sirolimus". Likely useless as predictor.

* **target**: The symbol of the gene that the comopund targets. For example "MTOR". Very useful, the actual target of the project.

* **moa**: A curated phrase representing the compound's mechanism of action. For example "MTOR inhibitor". Likely useful as a predictor.

* **canonical_smiles**: Canonical SMILES structure. An often long string of letters that describes a particular molecule in "1 dimension" rather than graphically. Essentially the "recipe" of a compound. The "canonical" indicates that the string has been generated by a procedure that always assigns the same SMILES structure to a particular compound. This is relevant because multiple SMILES strings can refer to the same molecule, for example "CCO" and "OCC" and "C(O)C" (ethanol). Likely extremely useful as a predictor. Can probably also be broken up into multiple sub-predictors. Perhaps one with the number of C's, one for the number of O's etc.

* **inchi_key**: InChIKey - hashed version of the InChi identifier (see http://www.iupac.org/inchi/). Might be useful.

* **compound_aliases**: Alternative names for the compound. Likely useless as predictor.
